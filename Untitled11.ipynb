# Importando bibliotecas e limpeza


```python
import pandas as pd #para ler, visualizar e printar infos do df
import matplotlib.pyplot as plt #para construir e customizar gráficos
import seaborn as sns #para visualizar uns gráficos
import numpy as np #numpy porque é sempre bom importar numpy né
```

```python
df = pd.read_csv("fipe_2022.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop_duplicates(keep = 'first', inplace=True)
```

```python
# Remover hífen e converter para inteiro na mesma coluna
df['fipe_code'] = df['fipe_code'].apply(lambda valor: int(valor.replace("-", "")))
```

```python
# Dicionário de mapeamento de meses para números
meses_para_numeros = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

# Substituir nomes dos meses por números inteiros
df['month_of_reference'] = df['month_of_reference'].map(meses_para_numeros)
```

```python
df.info()
```

```python
df.isnull().sum()
```

## Média de valor dos carros por mês 

```python
price_trends = df.groupby('month_of_reference')['avg_price_brl'].mean().reset_index()

_ = plt.figure(figsize=(12,8))
_ = plt.title("Média de preço por mês")
_ = plt.plot(price_trends['month_of_reference'], price_trends['avg_price_brl'], marker='o', color='red')
_ = plt.grid(True)
_ = plt.xlabel("Mês")
_ = plt.xticks(range(1,13))
_ = plt.ylabel("Média de valor em BRL")
```

## As 10 principais marcas e modelos de automóveis com preços médios mais altos


```python
# Calcula a média de preços para cada marca e modelo
brand_model_avg_prices = df.groupby(['brand', 'model'])['avg_price_brl'].mean().reset_index()

# Ordena as médias de preços em ordem decrescente e pega as top 10
top_10_high_prices = brand_model_avg_prices.nlargest(10, 'avg_price_brl')

# Cria um gráfico de barras para visualizar as top 10 médias mais altas
plt.figure(figsize=(12, 8))
bars = plt.barh(top_10_high_prices['brand'] + ' ' + top_10_high_prices['model'], top_10_high_prices['avg_price_brl'], color='red')
plt.title('Top 10 Car Brands and Models with Highest Average Prices')
plt.xlabel('Average Price (BRL)')
plt.ylabel('Brand and Model')
plt.gca().invert_yaxis()  # Inverte o eixo y para ter o maior valor no topo

# Adiciona os valores médios dentro das barras
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{bar.get_width():.2f}',
             va='center', ha='center', color='black', fontsize=10)

plt.tight_layout()
plt.show()
```

## Impacto do tipo de cambustível no valor médio 



```python
# Calculate the average prices for each fuel type
fuel_avg_prices = df.groupby('fuel')['avg_price_brl'].mean().reset_index()

# Create a bar plot to compare average prices based on fuel types
plt.figure(figsize=(10, 6))
plt.bar(fuel_avg_prices['fuel'], fuel_avg_prices['avg_price_brl'], color='purple')
plt.title('Average Car Prices Based on Fuel Types')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price (BRL)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Impacto do tipo de câmbio no valor médio 



```python
gear_avg_prices = df.groupby('gear')['avg_price_brl'].mean().reset_index()

# Create a grouped bar chart to examine gear preferences and their impact on prices
plt.figure(figsize=(10, 6))
bars1 = plt.bar(gear_avg_prices['gear'], gear_avg_prices['avg_price_brl'], color='blue')

plt.title('Gear Preferences and Average Car Prices')
plt.xlabel('Gear Type')
plt.ylabel('Average Price (BRL)')
plt.tight_layout()

# Add data labels to the bars
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 200, round(yval, 2), ha='center', color='black', fontsize=10)

plt.show()
```

## Diferença de valor médio entre carros novos e usados

```python
# Split data into brand new and used cars
new_cars = df[df['year_model'] == df['year_of_reference']]
used_cars = df[df['year_model'] != df['year_of_reference']]

# Calculate the average prices for brand new and used cars
avg_price_new_cars = new_cars['avg_price_brl'].mean()
avg_price_used_cars = used_cars['avg_price_brl'].mean()

# Create a bar plot to compare average prices of brand new and used cars
plt.figure(figsize=(8, 6))
plt.bar(['Brand New', 'Used'], [avg_price_new_cars, avg_price_used_cars], color=['blue', 'purple'])
plt.title('Average Prices of Brand New vs. Used Cars')
plt.xlabel('Car Condition')
plt.ylabel('Average Price (BRL)')
plt.tight_layout()
plt.show()
```

## Preço médio entre o top 5 marcas mais vendidas no Brasil

```python
# Filtrar para as marcas mais vendidas
top_brands = ['Fiat', 'GM - Chevrolet', 'VW - VolksWagen', 'Toyota', 'Hyundai']
df_top_brands = df[df['brand'].isin(top_brands)]

# Calcular o valor médio dos carros para cada marca
average_prices = df_top_brands.groupby('brand')['avg_price_brl'].mean()

# Criar um gráfico de barras
plt.figure(figsize=(15, 5))
average_prices.plot(kind='bar', color='blue')
plt.title('Average Car Prices of Top 5 Brands in 2022')
plt.xlabel('Brand')
plt.ylabel('Average Price (BRL)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Correlação entre tamanho do motor e preço médio

```python
# Calcular a correlação entre 'engine_size' e 'avg_price_brl'
correlation = df['engine_size'].corr(df['avg_price_brl'])

# Criar um gráfico de dispersão
plt.figure(figsize=(10, 6))
sns.scatterplot(x='engine_size', y='avg_price_brl', data=df)
plt.title('Scatter Plot of Engine Size vs Average Price')
plt.xlabel('Engine Size')
plt.ylabel('Average Price (BRL)')
plt.tight_layout()
plt.show()

print(f'Correlation between engine_size and avg_price_brl: {correlation:.2f}')
```

**Explorando a Relação Entre o Tamanho do Motor e os Preços dos Carros**

Nesta visualização, exploramos a possível conexão entre o tamanho do motor e os preços médios dos carros. O gráfico de dispersão representa o tamanho do motor de cada carro em relação ao seu preço médio correspondente. O coeficiente de correlação entre o tamanho do motor e o preço é calculado para quantificar a relação. Um valor de correlação próximo de 1 indica uma forte correlação positiva, o que sugere que motores maiores tendem a estar associados a preços mais altos. Por outro lado, um valor de correlação próximo de -1 sugere uma forte correlação negativa, indicando que motores maiores podem estar relacionados a preços mais baixos. Se o valor da correlação estiver próximo de 0, isso sugere uma correlação fraca ou negligenciável entre o tamanho do motor e o preço.

Compreender a interação entre o tamanho do motor e os preços pode fornecer insights sobre as preferências do consumidor, dinâmicas de mercado e estratégias de preços dentro da indústria automobilística.
